```python

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import requests

from datetime import datetime, timedelta
import os


# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode (no GUI)
chrome_options.binary_location = os.path.abspath("/usr/bin/google-chrome")

# Set up the Chrome driver using webdriver_manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)


BASE_DIR = os.path.join(os.path.dirname(os.path.abspath("__file__")), "work")
RAW_DIR = os.path.join(BASE_DIR, "Data", "Raw")
PAGE_DIR = os.path.join(BASE_DIR, "Data", "Pages")

initial_date = "2023-12-16"
end_date = "2024-03-23"
STEP_DAYS = 5

# Convert strings to datetime objects
start_date = datetime.strptime(initial_date, "%Y-%m-%d")
end_date = datetime.strptime(end_date, "%Y-%m-%d")

urls = []

# Generate URLs for each date in the range
current_date = start_date
while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    url = f"https://ambassadeurs.onlylyon.com/fr/agenda/past?published=1&happen_on={date_str}"
    urls.append((url, date_str))
    current_date += timedelta(days=STEP_DAYS)

# Scrape the data
data = []
for url, date in urls:
   
    driver.get(url)
    driver.get(10)
    
    html_content = driver.page_source

    soup = BeautifulSoup(html_content, 'html.parser')

    # data = soup.find(attrs={"data-tag-id": "agenda-publishing-occurrence-grid"})
    # print(data)
    with open(os.path.join(PAGE_DIR, f"{date}.html"), "w") as file:
        file.write(html_content)
    












```
